## Summary Statistics

```python
# Generate a (Pandas)summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.
#do 5 separate groupby's where each one is doing a different stat function like mean() end up with 5 different variables - can name them mean, median, variance, standard deviation, and SEM. You want to pull out just the Tumor Volume (mm3) column. These will be series.
# 5 different Groupby's with the appropriate accumulating functions 

grouped_mean = clean_pharmacy_study.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
grouped_mean

grouped_median = clean_pharmacy_study.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
grouped_median

grouped_var = clean_pharmacy_study.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
grouped_var

grouped_std = clean_pharmacy_study.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
grouped_std

grouped_sem = clean_pharmacy_study.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]
grouped_sem

#create a pd.DataFrame It will be a dictionary where you enter each column heading (mean, median, var, standard deviation,
#SEM) as the key and the valuse will be the stat series you just created. Dictionary -{ mean:grouped_mean, 
#median:grouped_median, variance: grouped_var, standard diviation:grouped_std, and SEM: grouped_sem  }
#value in the data fram is going to be one of the stat series you just calculated (grouped_mean etc. 
#will pull in those values)

pharmacy_stats_df = pd.DataFrame({"mean": grouped_mean, "median": grouped_median, "var": grouped_var, 
                                  "standard deviation": grouped_std, "SEM": grouped_sem})
pharmacy_stats_df
```

```python
#Generate a (Pyplot)summary statistics table of mean, median, variance, standard deviation, and SEM of tumoer
#volume by drug regimen using agg function

grouped_clean_pharmacy_study = clean_pharmacy_study.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median", "var", "std", "sem"]})
grouped_clean_pharmacy_study
#print(grouped_clean_pharmacy_study)
```
